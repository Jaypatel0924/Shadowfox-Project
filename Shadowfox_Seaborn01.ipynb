# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


var1=[1,2,3,4,5,6,7]
var2=[10,30,40,20,50,70,60]

x1=pd.DataFrame({"var1":var1, "var2":var2})

sns.lineplot(x="var1",y="var2",data=x1)
plt.show()



```

```python
# Using Matplotlib 

plt.plot(var1,var2)
plt.show()
```

# HeatMap


```python
import numpy as np
import pandas as pd

var1=np.linspace(1,20,20).reshape(4,5)
var1
```

```python
y={"fontsize":20,"color":"r"}
y1=sns.heatmap(var1,vmin=0,vmax=20,cmap="PuOr",annot=True
            ,annot_kws=y,linewidth=10,linecolor="grey"
            ,xticklabels=False,yticklabels=False)
y1.set(xlabel="Data",ylabel="Year",title="Hello Data")
#sns.set(font_scale=3)
plt.show()
```

# Pair Plot

```python
tips = sns.load_dataset("tips")
sns.pairplot(tips,hue="sex",kind="scatter")# kind are used for particular graph type
plt.show()
```

```python
sns.pairplot(tips,hue="sex",vars=["total_bill","tip"],hue_order=["Female","Male"])# vars are used to particular variable graph you want 
plt.show()
```

# Strip Plot

```python
m={"Male":">","Female":"*"}
tips = sns.load_dataset("tips")
sns.stripplot(x="day",y="total_bill",data=tips,hue="sex",linewidth=1,edgecolor="g",palette="Accent_r"
              ,alpha=0.9)
plt.legend(loc=2)
plt.show()
```

```python
sns.stripplot(y=tips["total_bill"])# Only a single variable plot
plt.show()
```

# Box Plot

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="tip",data=tips,hue="sex",palette='plasma',showmeans=True,linewidth=2,linecolor="black"
            ,meanprops={"marker":"*","markeredgecolor":"g"})
plt.show()
```

# Factor Plot Or Cat Plot

```python
sns.catplot(x="day",y="tip",data=tips,hue="sex",kind="boxen")
plt.show()
```

sns.catplot(x="day",y="tip",data=tips,hue="sex",kind="point",marker=[">","o"])
